```python
import sqlite3
import pandas as pd
data=pd.read_csv("C:/Users/hp/Documents/Getting Started/ImportCSV/OTT_Movies.csv",encoding="ISO-8859-1",index_col=[0])
data.head()
```

```python
data.columns=data.columns.str.strip()
data.columns=data.columns.str.replace(".","")
data.head()
```

```python
conn=sqlite3.connect('Movies.db')
c=conn.cursor()
data.to_sql('Movies_Data',conn,if_exists="replace")
```

```python
c.execute("Select * from Movies_Data;")
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

Showing all the Movies available on Amazon

```python
c.execute("Select * from Movies_Data where streaming_service='amazon';")
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data          
```

Showing all the Movies having genre as Romance and released between 2012 and 2021

```python
c.execute("Select * from Movies_Data where main_genre='romance' and (release_year between 2012 and 2021);")
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```

Showing the count of all genre types of Movies in descending order

```python
c.execute("Select main_genre, count(*) from Movies_Data group by main_genre order by count(*) desc;")
data=pd.DataFrame(c.fetchall())
data.columns=[x[0] for x in c.description]
data
```
